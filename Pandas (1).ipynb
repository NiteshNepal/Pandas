```python
#merging the two dataset 
series1={
    "Name":["x","y","z"],
    "USN":[1,2,3]
}
series2={
    "USN":[1,2,3],
    "Batch":["DS","AI","IOT"],
    "Age":[19,20,21]
}
```

```python
df1=pd.DataFrame(series1)
df2=pd.DataFrame(series2)
```

```python
df1.head()
```

```python
df2.head()
```

```python
pd.merge(df1,df2,on="USN")
```

```python
pd.merge(df1,df2,how="inner",on="USN")
```

```python
#merging the two dataset 
series1={
    "Name":["x","y","z"],
    "USN":[1,2,3]
}
series2={
    "USN":[1,2],
    "Batch":["DS","AI"],
    "Age":[19,20]
}
```

```python
df1=pd.DataFrame(series1)
df2=pd.DataFrame(series2)
```

```python
pd.merge(df1,df2,how="inner",on="USN")
```

```python
pd.merge(df1,df2,how="left",on="USN")  #all the rows in left and common.
```

```python
#merging the two dataset 
series1={
    "Name":["x","y"],
    "USN":[1,2]
}
series2={
    "USN":[1,2,3],
    "Batch":["DS","AI","IOT"],
    "Age":[19,20,18]
}
```

```python
df1=pd.DataFrame(series1)
df2=pd.DataFrame(series2)
```

```python
pd.merge(df1,df2,how="right",on="USN")  #all the rows in right and common.
```

```python
pd.merge(df1,df2,how="outer",on="USN")
```
