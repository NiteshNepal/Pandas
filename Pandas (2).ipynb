## iloc

 iloc is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

```python
#select all the rows in which Age is having null values.
x=df["Age"][df["Age"].isnull()].index
#x=np.where(df["Age"].isnull())
x
```

```python
"""iloc[x,y]===> iloc accepts two values.
x===> row index
y===>column index
"""
```

```python
df.iloc[x,:]
```

```python
df.iloc[x,2:6]
```

```python
#we can extract the maximum fare paid by the people.
max_fare=np.where(df["Fare"]==max(df["Fare"]))
```

```python
df.iloc[max_fare]
```

```python
df.iloc[1:100,1:4]
```

## Fillna operation

```python
df["Age"]=df["Age"].fillna(df["Age"].median())  #filling the missing place with the median
```

```python
df.isnull().sum()
```

```python
df["Fare"]=df["Fare"].fillna(min(df["Fare"])) #filling the missing value with the minimum value of the column.
```
