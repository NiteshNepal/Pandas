```python
d1={
    "Key":['ko','k8','k2'],
    "A":["A0","A1","A2"],
    "B":["B0","B1","B2"]
}
d2={
    "Key":['ko','k1','k2'],
    "C":["C0","C1","C2"],
    "D":["D0","D1","D2"]
    
}
```

```python
df1=pd.DataFrame(d1)
```

```python
df1
```

```python
df2=pd.DataFrame(d2)
```

```python
df2.head()
```

```python
pd.merge(df1,df2,on="Key")  #helps to merge two dataset based on common column "Key" which includes common rows only.
```

```python
pd.merge(df1, df2, on="Key", how="left")  
#"""here we can see that, the df1 is considered at first as we took the condition "left". 
#So k8 is not present in the df2 so the value of C and D columns in that particular row becomes NAN."""
```

```python
pd.merge(df1, df2, on="Key", how="right") #it is passed on the order df1 and then df2.
#"""here we can see that, the df2 is considered at first as we took the condition "right". 
#So k1 is not present in the df1 so the value of A and B columns in that particular row becomes NAN."""
```

```python
pd.merge(df2, df1, on="Key",how='inner') #only the common keys
```

```python
pd.merge(df2, df1, on="Key",how='outer') #all the values of the key.
```
